```python
import os
import gemini  # Replace with the actual module name if different

# Option 1: Pass the API key directly during model initialization
api_key = "API_KEY_HERE"  # Replace with your actual API key
model = gemini.load_model(api_key=api_key)

# Alternatively, if the API key should be set as an environment variable:
# os.environ["GEMINI_API_KEY"] = "AAPI_KEY_HERE"
# model = gemini.load_model()

print("Start chatting with Gemini! Type 'exit' to end the conversation.\n")

while True:
    user_input = input("You: ")
    
    # Check if the user wants to exit
    if user_input.strip().lower() == "exit":
        print("Conversation ended.")
        break
    
    # Get the response from the Gemini model (update this function if necessary)
    response = model.generate_response(user_input)
    
    # Output the response
    print("Gemini:", response, "\n")
```
